```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading the dataset
df = pd.read_csv('AirQuality.csv')
```

```python
df = pd.read_csv('AirQuality.csv',sep=";", decimal=",",header=0)
```

```python
print("No of rows in dataset:",df.shape[0])
print("No of columns in dataset:",df.shape[1])
```

```python
df.drop(columns=['Unnamed: 15', 'Unnamed: 16'],inplace=True)
```

```python
df.replace(to_replace=-200,value=np.nan,inplace=True)
```

```python
df.drop(columns=['NMHC(GT)'],inplace=True)
```

```python
col = ['CO(GT)', 'PT08.S1(CO)', 'C6H6(GT)','PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)','PT08.S5(O3)', 'T', 'RH', 'AH']
```

```python
for i in col:
    df[i] = df[i].fillna(df[i].mean())
```

```python
import seaborn as sns

plt.figure(figsize=(5,5))
sns.boxplot(data=df)
plt.xticks(rotation='vertical')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Replace 'df' with your DataFrame name
# If you're using Jupyter Notebook, you might not need to use plt.show() for inline plotting

# Example: Line plot of CO(GT) over time with a different color
df.plot(x='Date', y='CO(GT)', kind='line', color='green')  # Change color to green
plt.title('CO(GT) over Time')
plt.xlabel('Date')
plt.ylabel('CO(GT)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()

# Example: Scatter plot of RH vs Temperature (T) with a different color
df.plot(x='T', y='RH', kind='scatter', color='green')  # Change color to blue
plt.title('RH vs Temperature')
plt.xlabel('Temperature (T)')
plt.ylabel('Relative Humidity (RH)')
plt.show()

# Example: Histogram of C6H6(GT) with a different color
df['CO(GT)'].plot(kind='hist', bins=20, color='orange')  # Change color to orange
plt.title('Distribution of CO(GT)')
plt.xlabel('CO(GT)')
plt.ylabel('Frequency')
plt.show()
```
