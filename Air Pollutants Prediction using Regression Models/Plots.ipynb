```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('AirQuality.csv')
```

```python
df = pd.read_csv('AirQuality.csv',sep=";", decimal=",",header=0)
```

```python
print("No of rows in dataset:",df.shape[0])
print("No of columns in dataset:",df.shape[1])
```

```python
df.drop(columns=['Unnamed: 15', 'Unnamed: 16'],inplace=True)
```

```python
df.replace(to_replace=-200,value=np.nan,inplace=True)
```

```python
df.drop(columns=['NMHC(GT)'],inplace=True)
```

```python
col = ['CO(GT)', 'PT08.S1(CO)', 'C6H6(GT)','PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)','PT08.S5(O3)', 'T', 'RH', 'AH']
```

```python
for i in col:
    df[i] = df[i].fillna(df[i].mean())
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(data=df)
plt.xticks(rotation='vertical')
plt.show()
```

```python
df.columns
```

```python
df.plot(x='Date', y='CO(GT)', kind='line', color='green')  
plt.title('CO(GT) over Time')
plt.xlabel('Date')
plt.ylabel('CO(GT)')
plt.xticks(rotation=45)  
plt.tight_layout()  
plt.show()

df.plot(x='T', y='RH', kind='scatter', color='green') 
plt.title('RH vs Temperature')
plt.xlabel('Temperature (T)')
plt.ylabel('Relative Humidity (RH)')
plt.show()

df['CO(GT)'].plot(kind='hist', bins=20, color='orange')  
plt.title('Distribution of CO(GT)')
plt.xlabel('CO(GT)')
plt.ylabel('Frequency')
plt.show()
```

```python
non_null_counts = df.count()

plt.figure(figsize=(10, 6))  
non_null_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Values in Each Column')
plt.xlabel('Columns')
plt.ylabel('Number of Non-null Values')
plt.xticks(rotation=45) 
plt.tight_layout()  
plt.show()
```
